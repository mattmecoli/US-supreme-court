```python
import matplotlib.pyplot as plt
import plotly
import numpy as np
import pandas as pd
```

```python
orig = pd.read_csv('./output_files/OutcomeReport_ttt_FeatureImportInfoRBGinsburg.csv')
```

```python
orig.head()
```

```python
df = orig.drop(['Unnamed: 0', 'ID'], axis = 1)
```

```python
df.head()
```

```python
df = df.drop([1])
```

```python
df.head()
```

```python
df = df.transpose()
```

```python
df.columns = ['feature_importance']
```

```python
df = df.sort_values(by = ['feature_importance'], ascending = False)
```

```python
df.head(20)
```

```python
N = 20
features = df.index.values[0:N]
y_pos = np.arange(N)
imps = df.feature_importance.values[0:N]

plt.figure(figsize=(15,8))
plt.bar(y_pos, imps, align='center', alpha=0.5)
plt.xticks(y_pos, features, rotation = 90)
h = plt.ylabel('Feature Codebook Reference')
plt.title('Feature Importances for RBGinsburg')

plt.draw()
```
